```python
#Para instalar as bibliotecas necessárias
%pip install geopandas
%pip install mapclassify
# Importações das bibliotecas
# dados e geodados
import pandas as pd
import geopandas as gpd
# download de arquivos
import urllib.request
#visualização
import matplotlib.pyplot as plt
import folium
import mapclassify
#exibição dos gráficos no notebook
%matplotlib inline
```

```python
url = 'https://geoservicos.pbh.gov.br/geoserver/wfs?service=WFS&version=1.0.0&request=GetFeature&typeName=ide_bhgeo:ACADEMIA_CIDADE&srsName=EPSG:31983&outputFormat=application%2Fjson'
# baixe o arquivo da URL e salve com ’nome_do_arquivo.csv’
urllib.request.urlretrieve(url, 'aca_cidade.json')
df1 = gpd.read_file('aca_cidade.json')
df1
```

```python
url = 'https://geoservicos.pbh.gov.br/geoserver/wfs?service=WFS&version=1.0.0&request=GetFeature&typeName=ide_bhgeo:ACADEMIA_CEU_ABERTO&srsName=EPSG:31983&outputFormat=application%2Fjson'
# baixe o arquivo da URL e salve com ’nome_do_arquivo.csv’
urllib.request.urlretrieve(url, 'ceu_aberto.json')
df2 = gpd.read_file('ceu_aberto.json')
df2
```

```python
url = 'https://geoservicos.pbh.gov.br/geoserver/wfs?service=WFS&version=1.0.0&request=GetFeature&typeName=ide_bhgeo:BAIRRO_POPULAR&srsName=EPSG:31983&outputFormat=application%2Fjson'
# baixe o arquivo da URL e salve com ’nome_do_arquivo.csv’
urllib.request.urlretrieve(url, 'bairro_pop.json')
df3 = gpd.read_file('bairro_pop.json')
df3
```

```python
ceu_aberto_join = df1.sjoin(df=df3, how='left',
predicate='intersects',
lsuffix='left',
rsuffix='right')

aca_cidade_join = df2.sjoin(df=df3, how='left',
predicate='intersects',
lsuffix='left',
rsuffix='right')


merged = pd.concat([ceu_aberto_join, aca_cidade_join])
```

```python
aggregated = merged.groupby(['NOME_BAIRRO_POPULAR']).size().reset_index(name='quantidade')
sorted_df = aggregated.sort_values(by='quantidade', ascending=False)
top_15_bairros = sorted_df.head(15)

top_15_bairros
```

```python
df3.NOME = df3.NOME.str.upper()
sort = df3.merge(sorted_df, how  = 'left', left_on = 'NOME', right_on = 'NOME_BAIRRO_POPULAR')
sort.plot(column = 'quantidade', legend=True)
```
