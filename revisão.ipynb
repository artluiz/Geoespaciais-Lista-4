```python
arp = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\area_preservacao.geojson')
bah = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\bacia_hidrografica.geojson')
cur = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\curso_dagua.geojson')
es = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\estadosBR.geojson')
ma = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\macroregiao.geojson')
sede = gpd.read_file('.\\SIN492 - Datasets-20230614T222923Z-001\SIN492 - Datasets\\sede_municipal.geojson')
```

```python
sedes_juncao = sede.sjoin(df=bah, how='inner',
predicate='intersects',
lsuffix='left',
rsuffix='right')

bah_rio_doce = bah[bah['nome'] == 'Bacia do Rio Doce']

sedes_cortadas = gpd.clip(sedes_juncao, bah_rio_doce.geometry)

sedes_cortadas.shape[0]
```

```python
curso_juncao_in = cur.sjoin(df=arp, how='inner',
predicate='intersects',
lsuffix='left',
rsuffix='right')

curso_juncao_wi = cur.sjoin(df=arp, how='inner',
predicate='within',
lsuffix='left',
rsuffix='right')

merged = pd.concat([curso_juncao_in, curso_juncao_wi])

# Exibir os valores únicos da coluna "nome_left"
valores_unicos = merged['nome_left'].unique()
print(valores_unicos)
```
